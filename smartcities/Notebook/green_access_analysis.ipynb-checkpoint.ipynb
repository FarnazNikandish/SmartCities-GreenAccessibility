```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Folder path
data_path = "C:/Users/Farnaz/Desktop/smartcities/data/"

# Load GeoJSON
neighborhoods = gpd.read_file(data_path + "neighbourhoods.geojson")
greenspaces = gpd.read_file(data_path + "greenspaces.geojson")

# Check CRS
print("CRS neighborhoods:", neighborhoods.crs)
print("CRS greenspaces:", greenspaces.crs)

# Convert to a metric coordinate system (suitable for distance calculations)
neighborhoods = neighborhoods.to_crs(epsg=3857)
greenspaces = greenspaces.to_crs(epsg=3857)

# Calculate neighborhood and green space area (square kilometers)
neighborhoods["neigh_area_km2"] = neighborhoods.geometry.area / 1e6
greenspaces["green_area_km2"] = greenspaces.geometry.area / 1e6

# Calculate the centroid of each neighborhood
neighborhoods["centroid"] = neighborhoods.geometry.centroid

# Spatial join to find green spaces within each neighborhood
green_in_neigh = gpd.sjoin(greenspaces, neighborhoods, how="inner", predicate="intersects")

# Check the correct neighborhood column

# The column corresponding to the neighborhood is usually AREA_NAME_right
print(green_in_neigh.columns) 

# Sum the green space area in each neighborhood
green_area_sum = green_in_neigh.groupby("AREA_NAME_right")["green_area_km2"].sum()

# Add to the neighborhoods DataFrame
neighborhoods["green_area_km2"] = neighborhoods["AREA_NAME"].map(green_area_sum).fillna(0)

# Calculate the distance from each neighborhood to the nearest green space
def nearest_green_distance(point, greenspaces):
    return greenspaces.distance(point).min()

neighborhoods["dist_to_green"] = neighborhoods["centroid"].apply(
    lambda x: nearest_green_distance(x, greenspaces)
)
neighborhoods["dist_to_green_km"] = neighborhoods["dist_to_green"] / 1000

# Calculate the final Green Access Index (GAI)
# A combination of green space ratio and distance to green spaces
neighborhoods["GAI"] = (neighborhoods["green_area_km2"] / neighborhoods["neigh_area_km2"]) / (neighborhoods["dist_to_green_km"] + 0.1)

# Categorize access level (High / Medium / Low)
low = neighborhoods["GAI"].quantile(0.33)
high = neighborhoods["GAI"].quantile(0.66)

def categorize_gai(x):
    if x <= low:
        return "Low Access"
    elif x <= high:
        return "Medium Access"
    else:
        return "High Access"

neighborhoods["Access_Level"] = neighborhoods["GAI"].apply(categorize_gai)

# Check the number of neighborhoods in each category
print(neighborhoods["Access_Level"].value_counts())

# Display a map of the access index
fig, ax = plt.subplots(1,1, figsize=(12,12))
neighborhoods.plot(column="Access_Level",
                   legend=True,
                   linewidth=0.5,
                   edgecolor="black",
                   ax=ax)
greenspaces.plot(color="green", alpha=0.4, ax=ax)
ax.set_title("Green Access Level in Toronto Neighbourhoods (GAI Index)", fontsize=16)
ax.axis('off')
plt.show()
```

```python
import geopandas as gpd

# 300-meter buffer for green spaces
buffer_300 = greenspaces.buffer(300)

# Convert to GeoSeries
buffer_300 = gpd.GeoSeries(buffer_300, crs=greenspaces.crs)

# Overlay neighborhoods with the green space buffer
intersection = neighborhoods.geometry.intersection(buffer_300.union_all())

# Area of sections within the buffer (square meters)
neighborhoods["area_300"] = intersection.area

# Total area of the neighborhood
neighborhoods["area_total"] = neighborhoods.geometry.area

# Access coverage percentage
neighborhoods["GBC_300"] = (neighborhoods["area_300"] / neighborhoods["area_total"]) * 100

# Neighborhoods with the lowest access
neighborhoods[["AREA_NAME", "GBC_300"]].sort_values("GBC_300").head()
```
